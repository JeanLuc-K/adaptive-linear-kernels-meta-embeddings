```python
import numpy as np
import pandas as pd

from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import make_scorer, accuracy_score, f1_score
```

```python
# y is a dataframe with 1 column, convert to 1D array
y_raw = y.iloc[:, 0].astype(str)

# map labels
# usually ionosphere uses 'g' and 'b'
y_bin = y_raw.map({'g': 1, 'b': 0}).values

X_df = X.copy()

print("y labels:", np.unique(y_raw))
print("Converted y unique:", np.unique(y_bin))
```

```python
scoring = {
    "accuracy": make_scorer(accuracy_score),
    "f1": make_scorer(f1_score)
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

def evaluate_model(name, model, X, y):
    res = cross_validate(model, X, y, cv=cv, scoring=scoring, n_jobs=-1)
    print(f"\n{name}")
    print(f"  Accuracy: {res['test_accuracy'].mean():.4f} ± {res['test_accuracy'].std():.4f}")
    print(f"  F1-score:  {res['test_f1'].mean():.4f} ± {res['test_f1'].std():.4f}")
    return res
```

```python
linear_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="linear", C=1.0))
])

res_linear = evaluate_model("SVM (Linear Kernel)", linear_svm, X_df, y_bin)
```

```python
rbf_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
])

res_rbf = evaluate_model("SVM (RBF Kernel)", rbf_svm, X_df, y_bin)
```

```python
poly_svm = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="poly", degree=3, C=1.0, gamma="scale", coef0=1))
])

res_poly = evaluate_model("SVM (Polynomial Kernel, degree=3)", poly_svm, X_df, y_bin)
```

```python
from sklearn.metrics.pairwise import linear_kernel, rbf_kernel

def hybrid_kernel_matrix(X1, X2, alpha=0.5, gamma="scale"):
    # linear part
    K_lin = linear_kernel(X1, X2)

    # rbf part
    # choose gamma numeric
    if gamma == "scale":
        # same idea as sklearn: 1 / (n_features * X.var())
        g = 1.0 / (X1.shape[1] * np.var(X1))
    else:
        g = float(gamma)

    K_rbf = rbf_kernel(X1, X2, gamma=g)

    return alpha * K_lin + (1 - alpha) * K_rbf
```

```python
from sklearn.base import clone

def evaluate_precomputed_hybrid(alpha=0.5, C=1.0):
    accs, f1s = [], []

    X_scaled_np = StandardScaler().fit_transform(X_df)

    for train_idx, test_idx in cv.split(X_scaled_np, y_bin):
        X_train = X_scaled_np[train_idx]
        X_test = X_scaled_np[test_idx]
        y_train = y_bin[train_idx]
        y_test = y_bin[test_idx]

        K_train = hybrid_kernel_matrix(X_train, X_train, alpha=alpha, gamma="scale")
        K_test  = hybrid_kernel_matrix(X_test, X_train, alpha=alpha, gamma="scale")

        clf = SVC(kernel="precomputed", C=C)
        clf.fit(K_train, y_train)

        y_pred = clf.predict(K_test)

        accs.append(accuracy_score(y_test, y_pred))
        f1s.append(f1_score(y_test, y_pred))

    print(f"\nHybrid Kernel SVM (alpha={alpha}, C={C})")
    print(f"  Accuracy: {np.mean(accs):.4f} ± {np.std(accs):.4f}")
    print(f"  F1-score:  {np.mean(f1s):.4f} ± {np.std(f1s):.4f}")

evaluate_precomputed_hybrid(alpha=0.5, C=1.0)
```

```python
import numpy as np
import pandas as pd

X_df = X.copy()
y_raw = y.iloc[:, 0].astype(str)
y_bin = y_raw.map({'g': 1, 'b': 0}).values

print("X:", X_df.shape)
print("y counts (0=bad, 1=good):", np.bincount(y_bin))
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics.pairwise import linear_kernel, rbf_kernel

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

linear_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="linear", C=1.0))
])

rbf_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="rbf", C=1.0, gamma="scale"))
])

poly_model = Pipeline([
    ("scaler", StandardScaler()),
    ("svm", SVC(kernel="poly", degree=3, C=1.0, gamma="scale", coef0=1))
])

def hybrid_kernel_matrix(X1, X2, alpha=0.5, gamma="scale"):
    K_lin = linear_kernel(X1, X2)
    if gamma == "scale":
        g = 1.0 / (X1.shape[1] * np.var(X1))
    else:
        g = float(gamma)
    K_rbf = rbf_kernel(X1, X2, gamma=g)
    return alpha * K_lin + (1 - alpha) * K_rbf
```

```python
from sklearn.metrics import accuracy_score, f1_score

def oof_standard(model, X, y):
    y_pred = cross_val_predict(model, X, y, cv=cv, method="predict")
    y_score = cross_val_predict(model, X, y, cv=cv, method="decision_function")
    return y_pred, y_score

y_pred_lin,  y_score_lin  = oof_standard(linear_model, X_df, y_bin)
y_pred_rbf,  y_score_rbf  = oof_standard(rbf_model, X_df, y_bin)
y_pred_poly, y_score_poly = oof_standard(poly_model, X_df, y_bin)

def oof_hybrid(X_df, y, alpha=0.5, C=1.0):
    X_scaled = StandardScaler().fit_transform(X_df)
    y_pred = np.zeros_like(y)
    y_score = np.zeros_like(y, dtype=float)

    for train_idx, test_idx in cv.split(X_scaled, y):
        X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
        y_train = y[train_idx]

        K_train = hybrid_kernel_matrix(X_train, X_train, alpha=alpha, gamma="scale")
        K_test  = hybrid_kernel_matrix(X_test, X_train, alpha=alpha, gamma="scale")

        clf = SVC(kernel="precomputed", C=C)
        clf.fit(K_train, y_train)

        y_pred[test_idx] = clf.predict(K_test)
        y_score[test_idx] = clf.decision_function(K_test)

    return y_pred, y_score

# If you already found the best alpha in your sweep, put it here:
BEST_ALPHA = 0.5
y_pred_hyb, y_score_hyb = oof_hybrid(X_df, y_bin, alpha=BEST_ALPHA, C=1.0)

models = {
    "Linear": (y_pred_lin, y_score_lin),
    "RBF": (y_pred_rbf, y_score_rbf),
    "Poly (deg=3)": (y_pred_poly, y_score_poly),
    f"Hybrid (α={BEST_ALPHA})": (y_pred_hyb, y_score_hyb),
}

# Summary metrics
metrics = []
for name, (yp, _) in models.items():
    metrics.append({
        "Model": name,
        "Accuracy": accuracy_score(y_bin, yp),
        "F1": f1_score(y_bin, yp)
    })

df_metrics = pd.DataFrame(metrics).sort_values("F1", ascending=False)
display(df_metrics)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, auc
from sklearn.metrics import precision_recall_curve, average_precision_score

# Bar chart
plt.figure(figsize=(10, 4))
x = np.arange(len(df_metrics))
plt.bar(x - 0.2, df_metrics["Accuracy"], width=0.4, label="Accuracy")
plt.bar(x + 0.2, df_metrics["F1"], width=0.4, label="F1")
plt.xticks(x, df_metrics["Model"], rotation=15)
plt.ylim(0, 1.05)
plt.title("Model Comparison (Out-of-Fold CV)")
plt.legend()
plt.tight_layout()
plt.show()

# Confusion matrices
for name, (yp, _) in models.items():
    cm = confusion_matrix(y_bin, yp)
    disp = ConfusionMatrixDisplay(cm, display_labels=["Bad (0)", "Good (1)"])
    plt.figure(figsize=(4, 4))
    disp.plot(values_format="d")
    plt.title(f"Confusion Matrix: {name}")
    plt.tight_layout()
    plt.show()

# ROC
plt.figure(figsize=(6, 5))
for name, (_, ys) in models.items():
    fpr, tpr, _ = roc_curve(y_bin, ys)
    plt.plot(fpr, tpr, label=f"{name} (AUC={auc(fpr, tpr):.3f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves (Out-of-Fold CV)")
plt.legend()
plt.tight_layout()
plt.show()

# Precision Recall
plt.figure(figsize=(6, 5))
for name, (_, ys) in models.items():
    prec, rec, _ = precision_recall_curve(y_bin, ys)
    ap = average_precision_score(y_bin, ys)
    plt.plot(rec, prec, label=f"{name} (AP={ap:.3f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision–Recall Curves (Out-of-Fold CV)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.inspection import permutation_importance
from sklearn.metrics import make_scorer, f1_score

def perm_importance(model, X, y, name, n_repeats=15):
    scorer = make_scorer(f1_score)
    r = permutation_importance(model, X, y, scoring=scorer, n_repeats=n_repeats,
                               random_state=42, n_jobs=-1)

    imp = pd.DataFrame({
        "feature": X.columns,
        "importance_mean": r.importances_mean,
        "importance_std": r.importances_std
    }).sort_values("importance_mean", ascending=False)

    print(f"\nPermutation Importance (F1) for: {name}")
    display(imp.head(15))

    topk = 15
    dfp = imp.head(topk).iloc[::-1]
    plt.figure(figsize=(8, 6))
    plt.barh(dfp["feature"], dfp["importance_mean"])
    plt.title(f"Top {topk} Permutation Importances: {name}")
    plt.xlabel("Mean F1 decrease after shuffling feature")
    plt.tight_layout()
    plt.show()

    return imp

# Fit models on full data first
rbf_fit = rbf_model.fit(X_df, y_bin)
poly_fit = poly_model.fit(X_df, y_bin)

imp_linear = perm_importance(linear_fit, X_df, y_bin, "Linear SVM")
imp_rbf    = perm_importance(rbf_fit, X_df, y_bin, "RBF SVM")
imp_poly   = perm_importance(poly_fit, X_df, y_bin, "Poly SVM")
```

```python
def local_feature_impact(pipeline_model, X_df, y, sample_index=0, topk=12):
    scaler = pipeline_model.named_steps["scaler"]
    svm = pipeline_model.named_steps["svm"]

    x = X_df.iloc[[sample_index]]
    x_scaled = scaler.transform(x)

    base_score = svm.decision_function(x_scaled)[0]
    base_pred = int(base_score >= 0)

    deltas = []
    for j, fname in enumerate(X_df.columns):
        x_mod = x_scaled.copy()
        x_mod[0, j] = 0.0
        score_mod = svm.decision_function(x_mod)[0]
        deltas.append((fname, base_score - score_mod))

    df_local = pd.DataFrame(deltas, columns=["feature", "impact_on_score"])
    df_local["abs_impact"] = np.abs(df_local["impact_on_score"])
    df_local = df_local.sort_values("abs_impact", ascending=False)

    print("Sample index:", sample_index)
    print("True label:", y[sample_index], "(1=Good,0=Bad)")
    print("Pred label:", base_pred)
    print("Decision score:", base_score)

    display(df_local.head(topk))

    dft = df_local.head(topk).iloc[::-1]
    plt.figure(figsize=(8, 5))
    plt.barh(dft["feature"], dft["impact_on_score"])
    plt.title(f"Local Feature Impact for sample #{sample_index}")
    plt.xlabel("Positive supports predicted class, negative opposes it")
    plt.tight_layout()
    plt.show()

    return df_local

# Pick one correct Good and one correct Bad automatically (based on best model by F1)
best_model_name = df_metrics.iloc[0]["Model"]
best_pred, best_score = models[best_model_name]

good_indices = np.where((y_bin == 1) & (best_pred == 1))[0]
bad_indices  = np.where((y_bin == 0) & (best_pred == 0))[0]

idx_good = int(good_indices[0]) if len(good_indices) else 0
idx_bad  = int(bad_indices[0]) if len(bad_indices) else 0

print("Best model:", best_model_name)
print("Example Good index:", idx_good)
print("Example Bad index:", idx_bad)

# Only works directly for pipeline models (Linear/RBF/Poly). For Hybrid, use surrogate below.
local_feature_impact(linear_fit, X_df, y_bin, sample_index=idx_good)
local_feature_impact(linear_fit, X_df, y_bin, sample_index=idx_bad)
```

```python
def support_vector_stats(pipeline_model, name):
    svm = pipeline_model.named_steps["svm"]
    n_sv = svm.n_support_
    total_sv = n_sv.sum()
    print(f"\n{name}")
    print("Support vectors per class [Bad, Good]:", n_sv)
    print("Total support vectors:", total_sv)
    print("Fraction of dataset:", total_sv / len(X_df))

support_vector_stats(linear_fit, "Linear SVM")
support_vector_stats(rbf_fit, "RBF SVM")
support_vector_stats(poly_fit, "Poly SVM")
```
